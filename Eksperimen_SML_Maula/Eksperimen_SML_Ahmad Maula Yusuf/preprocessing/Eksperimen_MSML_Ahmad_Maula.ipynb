```python
# 5. Data Preprocessing

# Copy dataset
df_clean = df.copy()

# Hapus duplikasi
df_clean = df_clean.drop_duplicates()

# Pisahkan fitur dan target
X = df_clean.drop("target", axis=1)
y = df_clean["target"]

# Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dataset hasil preprocessing
df_processed = pd.DataFrame(X_scaled, columns=X.columns)
df_processed["target"] = y.values

df_processed.head()
```

```python
# Menyimpan dataset hasil preprocessing
os.makedirs("preprocessing/namadataset_preprocessing", exist_ok=True)

df_processed.to_csv(
    "preprocessing/namadataset_preprocessing/heart_processed.csv",
    index=False
)
```

```python
# =========================
# SKILLED  AUTOMATE PREPROCESSING
# =========================

%%writefile preprocessing/automate_Maula.py
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

def preprocess_data(
    input_path="namadataset_raw/heart.csv",
    output_path="preprocessing/namadataset_preprocessing/heart_processed.csv"
):
    df = pd.read_csv(input_path)

    # Drop duplicate data
    df = df.drop_duplicates()

    # Split features & target
    X = df.drop("target", axis=1)
    y = df["target"]

    # Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Combine processed data
    df_processed = pd.DataFrame(X_scaled, columns=X.columns)
    df_processed["target"] = y.values

    # Save processed dataset
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    df_processed.to_csv(output_path, index=False)

if __name__ == "__main__":
    preprocess_data()
```
